#### Financial Budget Analysis with Python
Each country has a financial budget that describes the government’s spending capacity in different sectors of the economy. You may get the task of analyzing a country’s financial budget every year if you are working as a data analyst in the media and communications field, as the media have to explain the government’s priorities for the complete financial year. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv")
data.head()
```
